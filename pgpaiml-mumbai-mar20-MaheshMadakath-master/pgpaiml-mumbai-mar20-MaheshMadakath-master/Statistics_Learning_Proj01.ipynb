### Part 1:Read the data set and replace dashes with 0 to make sure you can perform arithmetic  operations on the data (5 points)

```python
print(Laliga_raw) # Reading the data set
```

```python
Laliga = Laliga_raw.replace(to_replace = '-',value = 0) # Replacing "-" with 0
print(Laliga)  # Reading the data set revised
Laliga.info()  # Reading the data set 
# DataTypes is object eventhough data is numeric. Will not be able to perform arithmetic operation
```

```python
# converting object datatype carrying numerics to numeric datatype to be able to perform arithmetic operations
Laliga[['Points','GamesPlayed','GamesWon','GamesDrawn','GamesLost','GoalsFor','GoalsAgainst',
       'Champion','Runner-up','Third','Fourth','Fifth','Sixth','T']] = Laliga[['Points','GamesPlayed','GamesWon','GamesDrawn','GamesLost','GoalsFor','GoalsAgainst',
       'Champion','Runner-up','Third','Fourth','Fifth','Sixth','T']].apply(pd.to_numeric)
Laliga.info() # Recheck data type for numeric data
```

### Part 2: Print all the teams which have started playing between 1930-1980 (5 points)

```python
#In the data in Debut column we can see year in the format like 1930-31
#We extract first 4 to infer start Debut for the team
Laliga["Debut_Start"] = Laliga["Debut"].str.extract('(....)',expand=True) # extracting first 4 character
Laliga["Debut_Start"] = pd.to_numeric(Laliga["Debut_Start"]) # Converting to numeric to perform arithmetic operation
Team = Laliga[(Laliga["Debut_Start"] < 1980)& (Laliga["Debut_Start"] > 1930)] 
# Filtering Teams who had debut between 1930 and 1980 with 1930 and 1980 exclusive
print("There were " ,Team['Team'].nunique(), " Teams who had debut between 1930 and 1980 (exclusive 1930 and 1980)."
     "\n" 
     "Teams were:")
print(*Team['Team'].values, sep=" , ", end = ".")
```

### Part 3: Print the list of teams which came Top 5 in terms of points (2.5 points)

```python
Df = DataFrame({'Team': Laliga.sort_values('Points',ascending=False).head(5)['Team'], 
                'Points': Laliga.sort_values('Points',ascending=False).head(5)['Points']
                 }) # Sorting the Points in ascending and printing top 5 using head function
print("\u0332".join("Top Five Teams in terms of points were :"))
a=0
while(a < 5):
    print(Df.iloc[a,0],"\n" , "\t", "with",  Df.iloc[a,1] , "points")
    a = a + 1
# printing the top five teams with their points scored
```

### Part 4: Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)

```python
#defing required function
def Goal_diff_count(f,a):
    return f - a
#Using max and min function
Laliga['Goal_diff'] = Goal_diff_count(Laliga["GoalsFor"],Laliga["GoalsAgainst"])
print("Team '" ,Laliga[Laliga['Goal_diff'] == Laliga['Goal_diff'].max()].iloc[0,1], "' has the maximum goal difference.", "\n"
      "Team '" , Laliga[Laliga['Goal_diff'] == Laliga['Goal_diff'].min()].iloc[0,1], "' has the minimum goal difference." )
```

### Part 5: Create a new column with the name “Winning Percent” and append it to the data set (5  points)

```python
# Defining Winning Percentage function
def PercentWin(w,p):
    try:
        return round(((w * 100) / p),2)
    except ZeroDivisionError: # returing zero when denom is zero instead of error as data
        return 0.00   
# Creating new column and appending with data value calculated using Winning Percentage function
Laliga['WinningPercent'] = PercentWin(Laliga["GamesWon"], Laliga["GamesPlayed"] )
Laliga['WinningPercent'] = Laliga['WinningPercent'].fillna(0.00) #Last row contained Null value and filling with 0
print("\u0332".join("Top Five Teams in terms of Winning Percent are:"), 
      *Laliga.sort_values('WinningPercent',ascending=False).head(5)['Team'], sep = "\n") 
```
